```python
import pandas as pd
# Importa a biblioteca Pandas, que é essencial para trabalhar com estruturas de dados como DataFrames em Python.
# O 'as pd' é uma convenção padrão que permite referenciar a biblioteca de forma mais curta como 'pd' no código.
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv' 
# Atribui o endereço (URL) de um arquivo CSV à variável 'url'.

df_imigracao_canada = pd.read_csv(url) 
# Lê o arquivo CSV da 'url' e carrega seus dados em um DataFrame do pandas chamado 'dataframe_imigracao_canada'.

df_imigracao_canada.head() 
# Exibe as primeiras 5 linhas do DataFrame 'dataframe_imigracao_canada' para uma visualização inicial.
```

```python
df_imigracao_canada.info() 
# Exibe um resumo conciso do DataFrame 'dataframe_imigracao_canada'.
# Isso inclui o número de entradas, colunas, tipo de dado de cada coluna, e uso de memória.
```

```python
df_imigracao_canada.set_index('País', inplace=True) 
# Define a coluna 'País' como índice do DataFrame.
# O argumento 'inplace=True' modifica o DataFrame diretamente, sem a necessidade de reatribuir.
```

```python
df_imigracao_canada.head() 
# Este comando é usado para exibir as primeiras 5 linhas do DataFrame 'dataframe_imigracao_canada'.
# É útil para ter uma visão rápida da estrutura e do conteúdo dos dados.
```

```python
anos = list(map(str, range(1980, 2014))) 
# Gera uma sequência de números inteiros de 1980 a 2013 (inclusive), converte cada número para uma string e, em seguida, armazena essas strings em uma lista chamada 'anos'.

anos 
# Exibe o conteúdo da lista 'anos', mostrando todos os anos como strings.
```

```python
brasil = df_imigracao_canada.loc['Brasil', anos]
# Esta linha seleciona dados específicos do DataFrame 'dataframe_imigracao_canada' e os atribui à variável 'brasil'.

# Detalhes sobre o método '.loc':
# O '.loc' é um método do pandas usado para seleção de dados por rótulos (labels) de linhas e colunas.
# Ele permite selecionar linhas e/ou colunas usando seus nomes (rótulos), e não suas posições numéricas.
# A sintaxe geral é 'dataframe.loc[rótulo_linha, rótulo_coluna]'.

# No exemplo:
# 'Brasil': Este é o rótulo da linha que está sendo selecionada. Como o índice do 'dataframe_imigracao_canada' foi previamente definido como a coluna 'País', 'Brasil' é o nome da linha que contém os dados de imigração do Brasil.
# 'anos': Esta é uma lista de rótulos de colunas que estão sendo selecionadas. A variável 'anos' contém uma lista de strings representando os anos (ex: '1980', '1981', etc.). O '.loc' irá selecionar apenas as colunas correspondentes a esses anos para a linha 'Brasil'.

# Em resumo, 'brasil' se tornará uma Série (ou um pequeno DataFrame se 'anos' contivesse apenas uma coluna) contendo os dados de imigração do Brasil para cada um dos anos especificados na lista 'anos'.

brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
# Cria um dicionário chamado 'brasil_dict'.
# A chave 'ano' recebe uma lista com os rótulos do índice da série 'brasil' (que são os anos).
# A chave 'imigrantes' recebe uma lista com os valores da série 'brasil' (que são os números de imigrantes).

df_imigracao_brasil = pd.DataFrame(brasil_dict)
# Cria um novo DataFrame do pandas chamado 'dataframa_imigracao_brasil' usando o dicionário 'brasil_dict'.
# Isso transforma o dicionário em uma tabela com duas colunas: 'ano' e 'imigrantes'.

df_imigracao_brasil.head()
```

```python
import matplotlib.pyplot as plt
# Importa a biblioteca Matplotlib, que é amplamente utilizada para criar visualizações e gráficos em Python.
# O 'as plt' é uma convenção comum que permite referenciar a biblioteca de forma mais curta como 'plt' no código.
```

```python
plt.figure(figsize=(10,4))
# Cria uma nova figura e um conjunto de eixos, especificando o tamanho da figura como 10 polegadas de largura por 4 polegadas de altura.

plt.plot(df_imigracao_brasil['ano'], df_imigracao_brasil['imigrantes'])
# Cria um gráfico de linha.
# 'dataframa_imigracao_brasil['ano']' define os valores para o eixo X (os anos).
# 'dataframa_imigracao_brasil['imigrantes']' define os valores para o eixo Y (o número de imigrantes).

plt.title('Imigração do Brasil para o Canadá')
# Define o título principal do gráfico como 'Imigração do Brasil para o Canadá'.

plt.xlabel('Ano')
# Define o rótulo do eixo X (horizontal) como 'Ano'.

plt.ylabel('Número de imigrantes')
# Define o rótulo do eixo Y (vertical) como 'Número de imigrantes'.

plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
# Define os rótulos (ticks) específicos que aparecerão no eixo X do gráfico.
# Isso ajuda a tornar o eixo X mais legível, mostrando apenas anos selecionados em vez de todos os anos.

plt.show()
# Exibe o gráfico gerado.
# Sem este comando, o gráfico pode não aparecer em alguns ambientes de execução (como scripts Python puros).
```

```python
argentina = df_imigracao_canada.loc['Argentina', anos]
argentina_dict = {'ano':argentina.index.tolist(), 'imigrantes':argentina.values.tolist()}
df_imigracao_argentina = pd.DataFrame(argentina_dict)
df_imigracao_argentina.head()
```

```python
plt.figure(figsize=(12, 6))
# Cria uma nova figura (o "quadro" onde o gráfico será desenhado) e define seu tamanho.
# 'figsize=(12, 6)' significa 12 polegadas de largura por 6 polegadas de altura,
# tornando o gráfico mais largo para melhor visualização das duas linhas.

# Plotando os dados do Brasil
plt.plot(df_imigracao_brasil['ano'], df_imigracao_brasil['imigrantes'], color='green', label='Brasil')
# Plota a primeira linha do gráfico.
# 'df_imigracao_brasil['ano']' são os valores do eixo X (os anos).
# 'df_imigracao_brasil['imigrantes']' são os valores do eixo Y (o número de imigrantes do Brasil).
# 'color='green'' define a cor da linha como verde, conforme solicitado.
# 'label='Brasil'' define o rótulo para esta linha, que será usado na legenda do gráfico.

# Plotando os dados da Argentina
plt.plot(df_imigracao_argentina['ano'], df_imigracao_argentina['imigrantes'], color='blue', label='Argentina')
# Plota a segunda linha no MESMO gráfico.
# 'df_imigracao_argentina['ano']' são os valores do eixo X.
# 'df_imigracao_argentina['imigrantes']' são os valores do eixo Y (o número de imigrantes da Argentina).
# 'color='blue'' define a cor da linha como azul, conforme solicitado.
# 'label='Argentina'' define o rótulo para esta linha, que também será usado na legenda.

plt.title('Imigração Brasil e Argentina para o Canadá', fontsize=16)
# Define o título principal do gráfico, agora com o texto combinado.
# 'fontsize=16' aumenta o tamanho da fonte do título para melhor legibilidade.

plt.xlabel('Ano', fontsize=12)
# Define o rótulo do eixo X (horizontal) como 'Ano'.
# 'fontsize=12' aumenta o tamanho da fonte do rótulo do eixo.

plt.ylabel('Número de Imigrantes', fontsize=12)
# Define o rótulo do eixo Y (vertical) como 'Número de Imigrantes'.
# 'fontsize=12' aumenta o tamanho da fonte do rótulo do eixo.

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'], rotation=45)
# Define os rótulos específicos que aparecerão no eixo X para facilitar a leitura.
# 'rotation=45' gira os rótulos em 45 graus para evitar sobreposição, se houver muitos rótulos.

plt.yticks(fontsize=10)
# Define o tamanho da fonte dos rótulos do eixo Y para melhor legibilidade.

plt.grid(True, linestyle='--', alpha=0.7)
# Adiciona uma grade ao gráfico para facilitar a leitura dos valores.
# 'True' ativa a grade.
# 'linestyle='--'' define o estilo da linha da grade como tracejada.
# 'alpha=0.7' define a transparência da grade, para que ela não seja muito dominante.

plt.legend(fontsize=12)
# Exibe a legenda do gráfico. As legendas são criadas a partir dos 'label' que definimos nas chamadas 'plt.plot()'.
# Isso é crucial para identificar qual linha representa o Brasil e qual representa a Argentina.
# 'fontsize=12' aumenta o tamanho da fonte da legenda.

plt.tight_layout()
# Ajusta automaticamente os parâmetros do layout da figura para que todos os elementos (títulos, rótulos, legendas)
# caibam dentro da área da figura sem se sobreporem.

plt.show()
# Exibe o gráfico final com ambas as linhas, título, rótulos e legenda.
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# Esta linha é o método recomendado e mais flexível para criar gráficos no Matplotlib.
# 'plt.subplots()' cria uma figura (a janela ou tela do gráfico) e um conjunto de eixos (o "painel" onde os dados são plotados).
# Ele retorna dois objetos: 'fig' (a figura) e 'ax' (os eixos ou subgráfico).
# 'figsize=(8,4)' define o tamanho da figura (8 polegadas de largura por 4 de altura).
#
# POR QUE USAR ISSO AO INVÉS DE 'plt.figure()' e 'plt.plot()'?
# 1. Controle Explícito: Com 'fig' e 'ax', você tem controle explícito sobre a figura e os eixos.
#    Isso é especialmente útil quando você quer criar gráficos com múltiplos subplots (vários gráficos em uma única figura).
#    Ao invés de usar 'plt.plot()', 'plt.title()', etc., que operam no "gráfico atual" implicitamente,
#    você usa 'ax.plot()', 'ax.set_title()', etc., que operam diretamente nos eixos 'ax' que você definiu.
# 2. Melhor para Múltiplos Gráficos: Se você quiser plotar vários gráficos independentes ou organizá-los em uma grade,
#    'plt.subplots()' é a maneira mais eficiente e legível de fazer isso, pois você pode criar múltiplos 'ax' facilmente (ex: `fig, (ax1, ax2) = plt.subplots(1, 2)`).
# 3. Orientação a Objetos: É uma abordagem mais orientada a objetos, o que geralmente leva a um código mais robusto,
#    organizado e fácil de depurar para gráficos complexos. A maioria dos métodos de configuração (título, rótulos, limites)
#    passa de 'plt.' para 'ax.set_'.

ax.plot(df_imigracao_brasil['ano'], df_imigracao_brasil['imigrantes'])
# Plota a linha no objeto 'ax' (os eixos).
# 'df_imigracao_brasil['ano']' são os valores para o eixo X.
# 'df_imigracao_brasil['imigrantes']' são os valores para o eixo Y.
# A diferença para 'plt.plot()' é que aqui você especifica explicitamente em quais eixos a linha deve ser desenhada,
# dando mais controle em layouts com múltiplos subplots.

ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
# Define o título do gráfico para os eixos 'ax'.
# A principal diferença para 'plt.title()' é que 'ax.set_title()' opera diretamente nos eixos 'ax' específicos,
# enquanto 'plt.title()' opera no último ou "atual" gráfico criado. O '\n' cria uma quebra de linha no título.

ax.set_xlabel('Ano')
# Define o rótulo do eixo X (horizontal) para os eixos 'ax'.
# Equivalente a 'plt.xlabel()', mas aplicado ao objeto 'ax' explicitamente.

ax.set_ylabel('Número de Imigrantes')
# Define o rótulo do eixo Y (vertical) para os eixos 'ax'.
# Equivalente a 'plt.ylabel()', mas aplicado ao objeto 'ax' explicitamente.

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# Controla a localização dos ticks principais (marcações) no eixo X.
# 'ax.xaxis' refere-se ao objeto do eixo X.
# 'set_major_locator()' define como os ticks principais serão posicionados.
# 'plt.MultipleLocator(5)' faz com que os ticks principais apareçam a cada 5 unidades (neste caso, a cada 5 anos).
# Isso é útil para controlar a densidade dos rótulos e evitar sobreposição, tornando o gráfico mais limpo.
# É uma forma mais avançada e flexível de controlar os ticks em comparação com 'plt.xticks()' que apenas define os rótulos exatos.

plt.show()
# Exibe o gráfico gerado. Este comando é o mesmo e necessário independentemente da abordagem usada.
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))
# Cria uma figura (o "quadro") e uma matriz de eixos (subgráficos).
# '1, 2' significa 1 linha e 2 colunas de subgráficos.
# 'figsize=(15,5)' define o tamanho total da figura.

axs[0].plot(df_imigracao_brasil['ano'], df_imigracao_brasil['imigrantes'])
# Plota um gráfico de linha no primeiro subgráfico (à esquerda).

axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
# Define o título para o primeiro subgráfico.

axs[0].set_xlabel('Ano')
# Define o rótulo do eixo X para o primeiro subgráfico.

axs[0].set_ylabel('Número de Imigrantes')
# Define o rótulo do eixo Y para o primeiro subgráfico.

axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
# Define os ticks principais do eixo X para aparecerem a cada 5 anos no primeiro subgráfico.

axs[0].grid()
# Adiciona uma grade ao primeiro subgráfico.

axs[1].boxplot(df_imigracao_brasil['imigrantes'])
# Cria um gráfico de caixa (boxplot) dos dados de imigrantes do Brasil no segundo subgráfico (à direita).
# O boxplot mostra a distribuição dos dados (mediana, quartis e outliers).

axs[1].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
# Define o título para o segundo subgráfico.

axs[1].set_xlabel('Brasil')
# Define o rótulo do eixo X para o segundo subgráfico (aqui, representa a categoria "Brasil" para o boxplot).

axs[1].set_ylabel('Número de Imigrantes')
# Define o rótulo do eixo Y para o segundo subgráfico.

axs[1].grid()
# Adiciona uma grade ao segundo subgráfico.

plt.show()
# Exibe a figura contendo os dois subgráficos.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Dados fornecidos
lojas = ['A', 'B', 'C', 'D']
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# 2. Criar o DataFrame de vendas
# O dicionário 'vendas_2022' já tem os meses como chaves.
# Ao passar 'lojas' como índice, o pandas alinha as listas de vendas corretamente com cada loja.
df_vendas = pd.DataFrame(vendas_2022, index=lojas)

# Obter a lista de meses para usar no eixo X
meses = df_vendas.columns.tolist()

# 3. Criar a figura e os subplots (2 linhas, 2 colunas)
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # figsize ajusta o tamanho total da figura

# 4. Adicionar um título geral à figura
fig.suptitle('Variação de Vendas Mensais por Loja (2022)', fontsize=18, y=1.02) # y ajusta a posição vertical do título

# Achatando o array de eixos para facilitar a iteração
axs = axs.flatten()

# 5. Plotar os dados para cada loja em seu respectivo subplot
for i, loja in enumerate(lojas):
    # Seleciona os dados de vendas para a loja atual
    vendas_loja = df_vendas.loc[loja, meses]

    # Plota o gráfico de linha no subplot atual (axs[i])
    axs[i].plot(meses, vendas_loja, marker='o', linestyle='-') # Adiciona marcadores e estilo de linha

    # Configurações de cada subplot
    axs[i].set_title(f'Vendas da Loja {loja}', fontsize=14)
    axs[i].set_xlabel('Mês', fontsize=12)
    axs[i].set_ylabel('Número de Vendas', fontsize=12)

    # Rotacionar os rótulos do eixo X para melhor legibilidade
    axs[i].tick_params(axis='x', rotation=45)

    # Adicionar grade para facilitar a leitura dos valores
    axs[i].grid(True, linestyle='--', alpha=0.7)

# 6. Ajustar o layout para evitar sobreposição de elementos
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta o layout e reserva espaço para o suptitle

# 7. Exibir a figura
plt.show()
```
