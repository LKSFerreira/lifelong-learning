```python
import pandas as pd
import requests
import json

url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
resposta =  requests.get(url).text
dados_locacao_imoveis = json.loads(resposta)
dados_locacao_imoveis = pd.json_normalize(dados_locacao_imoveis, 'dados_locacao')
dados_locacao_imoveis.head()
```

```python
# Remover os dados em listas dentro do DataFrame;
dados_locacao_imoveis = dados_locacao_imoveis.explode(['datas_combinadas_pagamento','datas_de_pagamento','valor_aluguel'], ignore_index=True)
dados_locacao_imoveis.head()
```

```python
dados_locacao_imoveis.info()
```

```python
colunas_numericas = ['valor_aluguel']
```

```python
dados_locacao_imoveis[colunas_numericas] = dados_locacao_imoveis[colunas_numericas].replace(r'[\$]', '', regex=True).replace(r',', '.', regex=True).replace(r'[A-z]', '', regex=True).astype(float)
dados_locacao_imoveis.head()
```
